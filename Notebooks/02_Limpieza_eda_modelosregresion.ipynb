```python
df_Tata['date'] = pd.DatetimeIndex(pd.date_range('2021-11-20 03:00:00+05:30', periods=len(df_Tata),  freq='min')) #sobreescribo mi columna date con nuevo range por minutos 
df_Tata = df_Tata.set_index('date') # pongo la columna fecha como indice. para facilitar el analisis.
df_Tata.head()
```

```python
df_Tata['year'] = df_Tata.index.year
df_Tata['month'] = df_Tata.index.month
df_Tata['day'] = df_Tata.index.day
df_Tata['weekday'] = df_Tata.index.day_name()
df_Tata['weekofyear'] = df_Tata.index.isocalendar().week
df_Tata['hour'] = df_Tata.index.hour
df_Tata['minute'] = df_Tata.index.minute
```

```python
weekday_mapping = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}
df_Tata['weekday'] = df_Tata['weekday'].map(weekday_mapping)
```

```python
sumatorio_close_por_dia = df_Tata.groupby(['day']).agg({'close':'sum'})
sumatorio_close_por_dia['day'] = sumatorio_close_por_dia.index
# Creo otro df con la suma de los precios de cierre para cada dia. Osea la suma total del cierre de cada dia.
```

```python
sumatorio_close_por_dia.head()
```

```python
# evolucion de la suma total de "close" por dia.
plt.figure(figsize=(15,5))
sns.lineplot(data=sumatorio_close_por_dia, x=sumatorio_close_por_dia['day'], y=sumatorio_close_por_dia['close'])
```

```python
plt.figure(figsize = (9, 6))
pd.plotting.autocorrelation_plot(sumatorio_close_por_dia['close'])
print('Autocorrelation =', round(sumatorio_close_por_dia['close'].autocorr(), 4))
# vemos correla en el tiempo, a ver si hay patrones o dependencia temporal
# Por ejemplo, la autocorrelación con un retraso de 1 mide la correlación entre cada observación y la observación inmediatamente anterior. La autocorrelación con un retraso de 2 mide la correlación entre cada observación y la observación dos períodos de tiempo antes, y así sucesivamente
```

```python
plot_acf(sumatorio_close_por_dia['close']) # autocorrelacion (xt) respecto a (xt-1)
plt.grid()
plt.show()
plot_pacf(sumatorio_close_por_dia['close'], lags=14) # autocorrelacion parcial   (xt) (xt-h) h = valor rezagado aleatorio
plt.grid()
plt.show()
# esta correlacion nos muestra como correlan los precios de cierra respecto a los anteriores precios.
```

```python
precio_cierre_promedio = df_Tata['close'].resample('D').mean() # Creo precio-cierre-promedio como variable predictora.
rollingMEAN = precio_cierre_promedio.rolling(window=5).mean()
rollingSTD = precio_cierre_promedio.rolling(window=5).std()


fig, (ax1, ax2) = plt.subplots(2,1,figsize=(16,6))
plt.subplots_adjust(hspace=0.4)
ax1.plot(precio_cierre_promedio, c='green',label='close')
ax1.plot(rollingMEAN, c='blue', label='Rolling mean')
ax2.plot(rollingSTD, c='black',label = 'Rolling Std')

ax1.legend(fontsize=12), ax2.legend(fontsize=12)
ax1.set_ylabel('points'), ax2.set_ylabel('points')
ax1.margins(x=0), ax2.margins(x=0)
ax1.grid(), ax2.grid()
```
